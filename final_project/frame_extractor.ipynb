```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

video = "son.mp4"
# video = "hwang.mkv"
vidcap = cv2.VideoCapture(video)
vidcap
```

```python
count = 0

while(vidcap.isOpened()):
    vidcap.set(cv2.CAP_PROP_POS_FRAMES, count)
    ret, image = vidcap.read()
    
    cv2.imwrite("./son_images/frame_%d.jpg" % count, image) 
#     cv2.imwrite("./hwang_images/frame_%d.jpg" % count, image)
 
    print(f'Saved frame {count}' )
    count += 100
    
vidcap.release()
```
